```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
height=np.array([150,152,153,157,158,160,162,165,168,170,173,175,178,180,183,185,188,189,190]).reshape(-1,1)
```

```python
model=KMeans(n_clusters=3,random_state=42)
```

```python
model.fit(height)
```

```python
label=model.labels_
label
```

```python
centroid=model.cluster_centers_
centroid
```

```python
for i in range(len(height)):
    print(f"Height{height[i]}-->cm group:{label[i]}")
```

```python
plt.figure(figsize=(10,5))
plt.scatter(height,np.zeros_like(height),c=label,s=100,label="Data points")
plt.scatter(centroid,np.zeros_like(centroid),c='red',marker='X',s=200,label='centroid')
plt.title('K-means clustering of  student height')
plt.legend()
plt.grid()
plt.xlabel('Height(cm)')
plt.show()
```
