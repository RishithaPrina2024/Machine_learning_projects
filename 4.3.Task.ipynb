```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score,accuracy_score
```

```python
df=pd.read_csv('loan_approval_dataset.csv.csv')
df
```

```python
df['Loan_Approved']=df['Loan_Approved'].map({'Yes':1,'No':0})
```

```python
df.columns
```

```python
x=df.drop('Loan_Approved',axis=1)
y=df['Loan_Approved']
```

```python
num_col=x.select_dtypes(exclude="object").columns
cat_col=x.select_dtypes(include="object").columns
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
scaler=StandardScaler()
encoded=OneHotEncoder(sparse_output=False)
```

```python
xtrain[num_col]=scaler.fit_transform(xtrain[num_col])
xtest[num_col]=scaler.transform(xtest[num_col])
```

```python
xtrain_encoded=encoded.fit_transform(xtrain[cat_col])
xtest_encoded=encoded.transform(xtest[cat_col])
```

```python
xtrain.head()
```

```python
xtrain.drop(cat_col,axis=1,inplace=True)
xtest.drop(cat_col,axis=1,inplace=True)
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
ypred
```

```python
confusion_matrix(ytest,ypred)
```

based on these 50 is TN,45 is FP,45 is FN,60 is TP
