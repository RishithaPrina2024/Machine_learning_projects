```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score,accuracy_score
```

```python
df=pd.read_csv('loan_data.csv')
df
```

```python
df.shape
```

```python
df.columns
```

```python
x=df.drop('loan_status',axis=1)
y=df['loan_status']
```

```python
num_col=x.select_dtypes(exclude="object").columns
cat_col=x.select_dtypes(include="object").columns
```

```python
df[cat_col].columns
```

```python
x_train,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
scaler=StandardScaler()
encoded=OneHotEncoder(sparse_output=False)
```

```python
x_train[num_col]=scaler.fit_transform(x_train[num_col])
xtest[num_col]=scaler.transform(xtest[num_col])
```

```python
x_train_encoded=encoded.fit_transform(x_train[cat_col])
xtest_encoded=encoded.transform(xtest[cat_col])
```

```python
x_train.drop(cat_col,axis=1,inplace=True)
xtest.drop(cat_col,axis=1,inplace=True)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
ypred=model.predict(xtest)
ypred
```

```python
confusion_matrix(ytest,ypred)
```

```python
precision_score(ytest,ypred,pos_label=0)
```

```python
precision_score(ytest,ypred,pos_label=1)
```

```python
recall_score(ytest,ypred,pos_label=0)
```

```python
recall_score(ytest,ypred,pos_label=1)
```

```python
f1_score(ytest,ypred,pos_label=0)
```

```python
f1_score(ytest,ypred,pos_label=1)
```
